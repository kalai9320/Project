About Dataset
Data source: https://olympics.com/

Feature description:


1	year	The year of the Olympic Games

2	games_type	The type of games (summer/winter)

3	host_country	The name of the country that hosted the games

4	host_city	The name of the city that hosted the games

5	athletes	The number of athletes who participated in the games

6	teams	The number of teams who participated in the games

7	competitions	The number of competitions held during the games

8	country	The name of the country that participated in the games

9	gold	The number of gold medals won by the country

10	silver	The number of silver medals won by the country

11	bronze	The number of bronze medals won by the country

```python
data['year_hostcountry_hostcity_game']=data['year'].map(str)+' : '+data['host_country'].map(str)+' : '+data['host_city'].map(str)+ ' : '+data['games_type']
data['total']=data['gold']+data['silver']+data['bronze']
```

***How many olympics games have been held?***

```python
min=data.year.min()
max=data.year.max()
games_types=data.groupby('games_type').agg({'year':'nunique'})
games_types=pd.Series(games_types.year)
held_citys=data.groupby('games_type').agg({'host_city':'nunique'})
held_citys=pd.Series(held_citys.host_city)

print(f'olympics started from the year of {min} and till now {max} have been \n {games_types[0]} Summer Olympic Games held in {games_types[1]} cites \n {held_citys[0]} Winder Olympic Games held in {held_citys[1]} cites')
```

```python
athlets_anal_summer=data[data['games_type']=='Summer']
athlets_anal_summer=athlets_anal_summer.groupby('year_hostcountry_hostcity_game').agg({'athletes':'sum'})
athlets_anal_summer=pd.Series(athlets_anal_summer.athletes)
plt.plot(athlets_anal_summer,linestyle=':',linewidth=4,marker='8',markerfacecolor='r',markersize=10)
plt.rcParams['figure.figsize'] = [12, 12]
plt.xticks(rotation=270,ha='center')
plt.xlabel("summer olympic")
plt.ylabel("Athletes counts")
plt.ylim(bottom=2500,top=1073070)

plt.title("summer olympic Athletes counts",loc='center',size=30,color='r')
plt.legend(['summer olympic Athletes'])

for x,y in zip (athlets_anal_summer.index,athlets_anal_summer.values):
    Label='{:.0f}'.format(y)
    plt.annotate(Label,(x,y),textcoords='offset points',xytext=(0,10),ha='right')

plt.show()
```

```python
athlets_anal_winter=data[data['games_type']=='Winter']
athlets_anal_winter=athlets_anal_winter.groupby('year_hostcountry_hostcity_game').agg({'athletes':'sum'})
athlets_anal_winter=pd.Series(athlets_anal_winter.athletes)


plt.rcParams['figure.figsize'] = [12, 8]
plt.bar(athlets_anal_winter.index,athlets_anal_winter,color='g')
plt.xticks(rotation=270,ha='center')
plt.xlabel("Winter olympic")
plt.ylabel("Athletes counts")
plt.title("Winter olympic Athletes counts",loc='center',size=30,color='r')
plt.ylim(bottom=3500,top=90000)
plt.legend(['winter olympic Athletes couts'],loc='upper left',shadow=True,fancybox=True)
for x,y in zip (athlets_anal_winter.index,athlets_anal_winter.values):
    Label='{:.0f}'.format(y)
    plt.annotate(Label,(x,y),textcoords='offset points',xytext=(0,10),ha='center')
```

```python
Team_anal_summer=data[data['games_type']=='Summer']
Team_anal_summer=Team_anal_summer.groupby('year_hostcountry_hostcity_game').agg({'teams':'sum','competitions':'sum'})

plt.plot(Team_anal_summer,linestyle='dashdot',linewidth=3,marker='*',markerfacecolor='r',markersize=15,markeredgecolor='k')

plt.rcParams['figure.figsize'] = [12, 12]
plt.xticks(rotation=270,ha='center')
plt.xlabel("summer olympic")
plt.ylabel("teams and competitions counts")
plt.ylim(bottom=150,top=31600)

plt.title("summer olympic teams and competitions analysis",loc='center',size=25,color='g')
plt.legend(['teams','competitions'])

plt.show()
```

```python
Team_anal_Winter=data[data['games_type']=='Winter']
Team_anal_Winter=Team_anal_Winter.groupby('year_hostcountry_hostcity_game').agg({'teams':'sum','competitions':'sum'})
sns.lineplot(data=Team_anal_Winter,marker='s',markerfacecolor='g',markersize=10,linewidth=3,markeredgecolor='k')
plt.rcParams['figure.figsize'] = [12, 12]
plt.xticks(rotation=270,ha='center')
plt.xlabel("Winter olympic")
plt.ylabel("teams and competitions counts")
plt.ylim(bottom=140,top=3500)
plt.title("Winder olympic teams and competitions analysis",loc='center',size=25,color='g')
plt.show()
```
